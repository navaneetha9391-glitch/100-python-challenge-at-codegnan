```python
# FIND TOPPER
-  take no. of student as input
-  for each student 
  - take inputs as follows
      - name
      - marks  
```

```python
##### Who is topper
print("Welcome to the who is topper")
n = int(input("Enter the number of the students in class:"))
names = []
marks = []
class_marks = []
percentage = []
print("Enter the all the student marks")
for i in range (n) :
    print(f"Enter the roll numbers {i+1} student marks")
    name = input("Enter the student name:")
    print("Enter the Telugu, Hindi, English, Maths, Science, Social marks, as per the order")
    telugu, hindi, english, maths, science, social = map(int, input().split()[:6])
    total_marks = telugu + hindi + english + maths + science + social
    std_percentage = (total_marks/600) * 100
    # store all values in list
    names.append(name)
    marks.append(std_percentage)
    class_marks.append(total_marks)
    percentage.append(std_percentage)
# print the topper details
print("Topper in the class is")
top_marks = max(class_marks)
topper_pos = []
for i in range (n) :
    if class_marks[i] == top_marks:
        topper_pos.append(i)
#print the topper details
for i in range (len(toppers_pos)):
   print("Name of the student is:", names[i])
   print("marks of the student is:",names[i])
   print("total marks:",class_marks[i])
# Each student marks
for i in range(n):
    print("Name of the student is:",names[i], "and marks is:",marks[i], " and total marks:", class_marks)
    
    
```

- INPUT: [1,2,4,5,6,-5]

## 2.find the Largest value in list
- input: [1,2,4,5,6,-5]
- OUTPUT: 6
3. find the lenth of list
- input: [1,2,4,5,6,-5]
- output:6  


```python
numbers = [1, 2, 4, 5, 6, -5]

smallest = numbers[0]      # assume first element is smallest

for num in numbers:
    if num < smallest:
        smallest = num

print("Smallest value:", smallest)

largest = numbers[0]       # assume first element is largest

for num in numbers:
    if num > largest:
        largest = num

print("Largest value:", largest)
numbers = [1, 2, 4, 5, 6, -5]

count = 0

for _ in numbers:
    count = count + 1

print("Length of list:", count)
```

######  read list of numbers if numer is even add 0.5 and if the number is odd add one

```python
#read list of numbers if numer is even add 0.5 and if the number is odd add one
##lst=[1,2,3,4,5]
lst=list(map(int,input().split()))
for i in range(0,len(lst)):
    if lst[i]%2 :
        lst[i]=lst[i]+1
    else:
         lst[i]=lst[i]+0.5
print(lst)
```
